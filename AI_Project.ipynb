```
class HouseHoldObj_Dataset(utils.Dataset):

  def __init__(self, json_file_names):
    super().__init__()
    self.json_file_names = json_file_names

  def load_dataset(self, dataset_dir, annotations, is_train=True):
    self.classes = sorted(['candle', 'light', 'toothbrush', 'tv', 'bottle', 'phone', 'door', 'fan', 'table', 'chair'])
    self.json_ids = sorted(list(annotations.keys()))

    for idx, name in enumerate(self.classes):
      self.add_class('dataset', idx+1, name)
    
    for idx, filename in enumerate(list(self.json_file_names)):
      image_id = filename.split('/')[-1]
      try:
        annotation_id = [x for x in self.json_ids if image_id in x][0]
      except:
        print('Error loading image:',image_id)
        continue
      self.add_image('dataset'
                    , image_id=image_id
                    , path=filename
                    , class_id=[x for x in self.classes if x in filename][0]
                    , annotation = dict(annotations[self.json_ids[idx]])
                    )
      
  def extract_boxes(self, points):
    boxes = list()
    w = 1024
    h = 1024
    for point in points:
      point = point['shape_attributes']
      if point['name'] == 'ellipse':
        col_start = point['cx'] - point['rx']
        row_start = point['cy'] - point['ry']
        col_end = point['rx']*2
        row_end = point['ry']*2
      elif point['name'] == 'rect':
        col_start = point['x']
        row_start = point['y']
        col_end = point['width'] + col_start
        row_end = point['height'] + row_start
      else:
        col_start = min(point['all_points_x'])
        col_end = max(point['all_points_x'])
        row_start = min(point['all_points_y'])
        row_end = max(point['all_points_y'])
      boxes.append((col_start, row_start, col_end, row_end))
    return boxes, w, h
  
  def load_mask(self, image_id):
    info = self.image_info[image_id]
    path = info['annotation']
    boxes, w, h = self.extract_boxes(path['regions'])
    masks = np.zeros([h, w, len(boxes)], dtype=np.uint8)
    class_ids = list()
    for idx, box in enumerate(boxes):
      col_start, row_start, col_end, row_end = box
      masks[int(row_start):int(row_end),int(col_start):int(col_end),idx] = 1
      class_ids.append(self.classes.index(info['class_id'])+1)
    return masks, np.array(class_ids, dtype=np.uint8)
  
  def image_reference(self, image_id):
    info = self.image_info[image_id]
    if info['source'] == 'dataset':
      return info['dataset']
    else:
      super(self.__class__).image_reference(self, image_id)

```
